```python
# prompt: armar un dataframe de 100 filas con datos random con apellido y nombre, numero de dni, materias de educacion nivel secundario de la argentina, que contenga las sigientes columnas. materia escolar, nota 1er trimestre, nota 2do trimestre, nota 3er trimestre, faltas a clases

import pandas as pd
import numpy as np
import random

# Define materias
materias = [
    "Matemática", "Lengua y Literatura", "Historia", "Geografía", "Ciencias Naturales",
    "Educación Física", "Música", "Artes Visuales", "Inglés", "Ciudadanía y Participación",
    "Tecnología", "Prácticas del Lenguaje"
]

# Crear lista de apellidos y nombres (reemplazar con una lista más completa si es necesario)
apellidos = ["García", "Rodríguez", "González", "Fernández", "López", "Martínez", "Sánchez", "Pérez"]
nombres = ["Juan", "Ana", "Pedro", "María", "Luis", "Laura", "Carlos", "Sofía"]

# Función para generar un DNI aleatorio
def generar_dni():
  return random.randint(10000000, 99999999)


# Crear el DataFrame
data = {
    'Apellido': [random.choice(apellidos) for _ in range(100)],
    'Nombre': [random.choice(nombres) for _ in range(100)],
    'DNI': [generar_dni() for _ in range(100)],
    'Materia Escolar': [random.choice(materias) for _ in range(100)],
    'Nota 1er Trimestre': np.random.randint(1, 11, size=100),
    'Nota 2do Trimestre': np.random.randint(1, 11, size=100),
    'Nota 3er Trimestre': np.random.randint(1, 11, size=100),
    'Faltas a Clases': np.random.randint(0, 20, size=100)  # Faltas hasta 20
}

df = pd.DataFrame(data)
df
```

```python
# prompt: generar un gráfico de barras con el promedio general de cada materia, ordenado de mayor a menor, agregar las faltas

import matplotlib.pyplot as plt

# Calcular el promedio general de cada materia
promedios_materias = df.groupby('Materia Escolar')[['Nota 1er Trimestre', 'Nota 2do Trimestre', 'Nota 3er Trimestre']].mean().mean(axis=1)

# Ordenar los promedios de mayor a menor
promedios_ordenados = promedios_materias.sort_values(ascending=False)

# Calcular la cantidad total de faltas por materia
faltas_por_materia = df.groupby('Materia Escolar')['Faltas a Clases'].sum()

# Crear el gráfico de barras
plt.figure(figsize=(14, 6))
plt.bar(promedios_ordenados.index, promedios_ordenados.values)
plt.xticks(rotation=90)
plt.xlabel('Materias')
plt.ylabel('Promedio General')
plt.title('Promedio General por Materia (Ordenado de Mayor a Menor)')

# Agregar las faltas como etiquetas en las barras
for i, v in enumerate(promedios_ordenados):
    plt.text(i, v + 0.1, f"Faltas: {faltas_por_materia[promedios_ordenados.index[i]]}", ha='center')

plt.tight_layout()
plt.show()
```
